Import all the package that we will use for the code.

```
# Import the necessary packages
import argparse
import cv2
from google.colab.patches import cv2_imshow
```

Set the path for the image that we will use for testing the code.

When you mount the code, folder **"drive"** will appear on the left screen when you click the **"file"** symbol. Then you just need to find the image in your drive and right click on that image.

A pop up will show some list then just click **"copy path"**. After that you just need to paste the path here.

```
# Construct the argument parser and parse the arguments
args = {
	# Task 1
	"image": '/content/drive/My Drive/Colab Notebooks/Image-Processing/corona.png',
	"image2": '/content/drive/My Drive/Colab Notebooks/Image-Processing/flower.jpg'
}
```

Now that we have the path to the image, we can load it off the disk using the cv2.imread function on Line 1. The cv2.imread function returns a NumPy array representing the image.

Then, displaying the actual image on our screen. The parameter is a reference to the image we loaded before.

```
# Task 2
# Load the image and show it here :
image = cv2.imread(args['image2'])
cv2_imshow(image)
```

Flipping an image is accomplished by making a call to the cv2.flip function on Line 2. The cv2.flipmethod requires two arguments: the image we want to ﬂip and a ﬂip code that is used to determine how we are going to ﬂip the image.

Using a ﬂip code value of 1 indicates that we are going to ﬂip the image horizontally, around the y-axis (Line 2).

```
# Task 3
# Flip the image horizontally and show the image
flippedy = cv2.flip(image,1)

cv2_imshow(image)
cv2_imshow(flippedy)
```

Specifying a ﬂip code of 0 indicates that we want to ﬂip the image vertically, around the x-axis (Line 2).

```
# Task 4
# Flip the image vertically and show the image
flippedx = cv2.flip(image,0)
cv2_imshow(image)
print('---------------------------------')
cv2_imshow(flippedx)
```

Finally, using a negative ﬂip code (Line 2) ﬂips the image around both axes.

```
# Task 5
# Flip the image along both axes and show the image
flippedboth = cv2.flip(image,-1)
cv2_imshow(image)
print('---------------------------------')
cv2_imshow(flippedboth)
```
