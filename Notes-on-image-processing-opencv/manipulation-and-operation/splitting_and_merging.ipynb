Now that we have the path to the image, we can load it off the disk using the cv2.imread function on Line 5. The cv2.imread function returns a NumPy array representing the image.

Then, splitting the channels using a call cv2.split on Line 6.

Normally, we think of images in the RGB color space – the red pixel ﬁrst, the green pixel second, and the blue pixel third. However, OpenCV stores RGB images as NumPy arrays in reverse channel order. Instead of storing an image in RGB order, it instead stores the image in BGR order; thus we unpack the tuple in reverse order. Lines 9-12 show each channel individually.

```
# Load the image and grab each channel: Red, Green, and Blue
# NOTE: OpenCV stores an image as NumPy array with its
# channels in reverse order! When we call cv2.split, we are
# actually getting the channels as Blue, Green, Red!
# Task 2
# Load the image here :
image = cv2.imread(args["image2"])

image = (B, G, R) = cv2.split(image)
# Task 3
# Show each channel individually :
cv2_imshow(R)
cv2_imshow(G)
cv2_imshow(B)
```

We can also merge the channels back together again using the cv2.merge function. We simply specify our channels, again in BGR order, and then cv2.merge takes care of the rest for us (Line 2).

```
# Task 4
# Merge the image back together again using opencv function and show the image
merged = cv2.merge((B, G, R))
cv2_imshow(merged)
```

On Line 2 we construct a NumPy array of zeros, with the same width and height as our original image. Then, in order to construct the Red channel representation of the image, we make a call to cv2.merge, but specifying our zeros array for the Green and Blue channels. We take similar approaches to the other channels in Line 4 and 5.

```
# Now, let's visualize each channel in color
image = cv2.imread(args["image2"])
zeros = np.zeros(image.shape[:2], dtype = "uint8")
cv2_imshow(cv2.merge([zeros, zeros, R]))
cv2_imshow(cv2.merge([zeros, G, zeros]))
cv2_imshow(cv2.merge([B, zeros, zeros]))
```
