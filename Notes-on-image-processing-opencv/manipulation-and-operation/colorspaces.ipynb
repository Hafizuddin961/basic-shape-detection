Import all the package that we will use for the code.

```
# Import the necessary packages
import numpy as np
import argparse
import cv2
from google.colab.patches import cv2_imshow
```

Set the path for the image that we will use for testing the code.

```
# Construct the argument parser and parse the arguments
args = {
	# Task 1
	"image": '/content/drive/My Drive/Colab Notebooks/Image-Processing/corona.png',
	"image2": '/content/drive/My Drive/Colab Notebooks/Image-Processing/flower.jpg'
	
}
```

Now that we have the path to the image, we can load it off the disk using the cv2.imread function on Line 1. The cv2.imread function returns a NumPy array representing the image.

Then, displaying the actual image on our screen. The parameter is a reference to the image we loaded before.

```
# Task 2
# Load the image and show it here :
image = cv2.imread(args["image2"])
print(image.shape)
cv2_imshow(image)
```

On Line 2, we convert our image from the RGB color space to grayscale by specifying the cv2.COLOR_BGR2GRAY ﬂag.

```
# Task 3
# Convert the image to grayscale and show the image
print(image.shape)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

Converting our image to the HSV color space is performed on Line 3 by specifying the cv2.COLOR_BGR2HSV ﬂag.

```
# Task 4
# Convert the image to the HSV (Hue, Saturation, Value) color spaces and show the image :
HSV = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
cv2_imshow(HSV)
```

Finally, on Line 2, we convert to the L*a*b* color space by using the cv2.COLOR_BGR2LAB ﬂag.

```
# Task 5
# Convert the image to the L*a*b* color spaces and show the image
LAB = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
cv2_imshow(LAB)
```
