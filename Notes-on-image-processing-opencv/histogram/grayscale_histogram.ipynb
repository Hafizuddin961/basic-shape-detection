Now that we have the path to the image, we can load it off the disk using the cv2.imread function on Line 1. The cv2.imread function returns a NumPy array representing the image.

Then, convert the image to grayscale image

Finally, displaying the actual image on our screen. The parameter is a reference to the image we convert before.

```
# Task 2
# Load the image and show it here :
image = cv2.imread(args["image"])
cv2_imshow(image)

# Task 3
# Convert image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Task 4
# Show the image
cv2_imshow(gray)
```

Line 2 computes the actual histogram. Go ahead and match the arguments of the code up with the function documentation above. We can see that our ﬁrst parameter is the grayscale image. A grayscale image has only one channel, hence we have a value of [0] for channels. We don’t have a mask, so we set the mask value to None. We will use 256 bins in our histogram, and the possible values range from 0 to 256.

Finally, a call plt.plot() to plots our grayscale histogram.

```
# Construct a grayscale histogram
hist = cv2.calcHist([gray], [0], None, [256], [0, 256])

# Plot the histogram
plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("Bins")
plt.ylabel("# of Pixels")
plt.plot(hist)
plt.xlim([0, 256])

# Task 5
# Show the plot here :
```

```
import cv2
import numpy as np
from matplotlib import pyplot as plt

color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([image],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])

plt.title("'Flattened' Color Histogram")
plt.xlabel("Bins")
plt.ylabel("# of Pixels")
plt.show()
```
