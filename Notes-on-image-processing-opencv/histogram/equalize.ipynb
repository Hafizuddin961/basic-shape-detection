## **In this code, we are going to do histogram equalization. Histogram equalization improves the contrast of an image by “stretching” the distribution of pixels.**

Import all the package that we will use for the code.

```
# Import the necessary packages
import numpy as np
import argparse
import cv2
from google.colab.patches import cv2_imshow
```

Set the path for the image that we will use for testing the code.

When you mount the code, folder **"drive"** will appear on the left screen when you click the **"file"** symbol. Then you just need to find the image in your drive and right click on that image.

A pop up will show some list then just click **"copy path"**. After that you just need to paste the path here.

```
# Construct the argument parser and parse the arguments
args = {
	# Task 1
	"image": "/content/drive/My Drive/Colab Notebooks/Image-Processing/flower.jpg"
}
```

Now that we have the path to the image, we can load it off the disk using the cv2.imread function on Line 2. The cv2.imread function returns a NumPy array representing the image.

Then, convert the image into grayscale.

```
# Task 2
# Load the image and show it here :
image = cv2.imread(args["image"])
cv2_imshow(image)

# Task 3
# Convert image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Task 4
# Show the image
cv2_imshow(gray)
```

Performing histogram equalization is done using just a single function: cv2.equalizeHist, which accepts a single parameter, the grayscale image we want to perform histogram equalization on.

```
# Apply histogram equalization to stretch the constrast
# of our image
eq = cv2.equalizeHist(gray)
```

Showing the image.

```
# Show our images -- notice how the constrast of the second
# image has been stretched
cv2_imshow(np.hstack([gray, eq]))
```
